```python
import pandas as pd
```

```python
data = pd.read_csv('finished_distance_data.csv')
print(data.shape)
data.head()
```

```python
remove_nan= data.dropna(subset=["total_nodes"])
remove_nan.shape
```

```python
remove_nan.head()
```

```python
for harpnum in remove_nan.loc[:,"HARPNUM"]:
    avg_x = remove_nan.loc[remove_nan["HARPNUM"]==harpnum, "hc_x"].var()
    avg_y = remove_nan.loc[remove_nan["HARPNUM"]==harpnum, "hc_y"].var()
    avg_z = remove_nan.loc[remove_nan["HARPNUM"]==harpnum, "hc_z"].var()
    remove_nan.loc[remove_nan["HARPNUM"]==harpnum, "mean_location_variance"] = (avg_x+avg_y+avg_z)/3.
    
```

```python
test = remove_nan.dropna(subset=["mean_location_variance"])
test.head()
test.loc[:, "mean_location_variance"].mean()
```

```python
mean_count = 0
mean_val = 0
for harpnum in remove_nan.loc[:, "HARPNUM"]:
    mean_count += 1
    mean_val += remove_nan.loc[remove_nan["HARPNUM"]==harpnum].shape[0]
    
print(mean_val/mean_count)
    
```
